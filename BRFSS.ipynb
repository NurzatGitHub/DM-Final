```python
import pandas as pd

file_path = r"C:\AI\DM\DM-Final\2015.csv"
brfss = pd.read_csv(file_path)
pd.set_option('display.max_columns', 500)
```

```python
columns_selected = [
    '_MICHD', '_RFHYPE5', 'TOLDHI2', '_CHOLCHK', '_BMI5', 'SMOKE100',
    'CVDSTRK3', 'DIABETE3', '_TOTINDA', '_FRTLT1', '_VEGLT1', '_RFDRHV5',
    'HLTHPLN1', 'MEDCOST', 'GENHLTH', 'MENTHLTH', 'PHYSHLTH', 'DIFFWALK',
    'SEX', '_AGEG5YR', 'EDUCA', 'INCOME2'
]

brfss = brfss[columns_selected]
```

```python
def clean_binary(df, col, mapping, remove_values=None):
    df[col] = df[col].replace(mapping)
    if remove_values:
        df = df[~df[col].isin(remove_values)]
    return df
```

```python
brfss = clean_binary(brfss, '_MICHD', {2:0})
brfss = clean_binary(brfss, '_RFHYPE5', {1:0, 2:1}, remove_values=[9])
brfss = clean_binary(brfss, 'TOLDHI2', {2:0}, remove_values=[7,9])
brfss = clean_binary(brfss, '_CHOLCHK', {2:0, 3:0}, remove_values=[9])
brfss['_BMI5'] = (brfss['_BMI5'] / 100).round(0)
brfss = clean_binary(brfss, 'SMOKE100', {2:0}, remove_values=[7,9])
brfss = clean_binary(brfss, 'CVDSTRK3', {2:0}, remove_values=[7,9])

brfss['DIABETE3'] = brfss['DIABETE3'].replace({2:0, 3:0, 1:2, 4:1})
brfss = brfss[~brfss['DIABETE3'].isin([7,9])]
brfss = clean_binary(brfss, '_TOTINDA', {2:0}, remove_values=[9])
brfss = clean_binary(brfss, '_FRTLT1', {2:0}, remove_values=[9])
brfss = clean_binary(brfss, '_VEGLT1', {2:0}, remove_values=[9])
brfss = clean_binary(brfss, '_RFDRHV5', {1:0, 2:1}, remove_values=[9])
brfss = clean_binary(brfss, 'HLTHPLN1', {2:0}, remove_values=[7,9])
brfss = clean_binary(brfss, 'MEDCOST', {2:0}, remove_values=[7,9])
brfss = brfss[~brfss['GENHLTH'].isin([7,9])]
brfss['MENTHLTH'] = brfss['MENTHLTH'].replace({88:0})
brfss = brfss[~brfss['MENTHLTH'].isin([77,99])]
```

```python
brfss = brfss.dropna()
```

```python
output_file = 'brfss_2015_cleaned.csv'
brfss.to_csv(output_file, index=False)

print(f"save file {output_file}")
print(brfss.shape)
brfss.head()
```
